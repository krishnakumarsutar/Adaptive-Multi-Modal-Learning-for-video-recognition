## Forming txt files for reference to video directory and label associated with it. 

```python
import os
```

```python
rootdir = 'kinetics400train'
class_dict = {}
classes = []
for file in os.listdir(rootdir):
    #file - abselling,..etc.
    classes.append(file) if file not in classes else classes
    d = os.path.join(rootdir, file)
    for subfile in os.listdir(d):
        class_dict[os.path.splitext(os.path.basename(os.path.join(d,subfile)))[0]] = classes.index(file)
```

```python
rootdir = 'kinetics400train_images'
with open('train.txt', 'w') as f:
    for file in os.listdir(rootdir):
        #path to video folder
        d = os.path.join(rootdir, file)
        if os.path.isdir(d):
            #st = d + ' ' + '1 ' + str(len([name for name in os.listdir(d)])) + ' ' + str(classes.index(class_dict[file]))
            #f.write(st)
            print(d,1,len([name for name in os.listdir(d)]),class_dict.get(file),file=f)
```

```python
rootdir = 'kinetics400train_audio'
with open('train.txt', 'a') as f:
    for file in os.listdir(rootdir):
        #path to video folder
        d = os.path.join(rootdir, file)
        print(d,1,len([name for name in os.listdir('kinetics400train_images/'+file[0:-4])] if os.path.exists('kinetics400train_images/'+file[0:-4]) else "5656565656"),class_dict.get(file[0:-4]),file=f)
```

```python
with open('sampledimages.txt', 'r') as f:
    lines = f.readlines()

# remove spaces
lines = [line.replace('  ', ' ') for line in lines]
lines = [line.replace('[', '') for line in lines]
lines = [line.replace(']', '') for line in lines]

# finally, write lines in the file
with open('sampledimages.txt', 'w') as f:
    f.writelines(lines)
```

```python
frames = []
with open('sampledimages.txt', 'r') as f:
  line = f.readlines()
line = [lines.replace('\n', '') for lines in line]
for l in line:
  elements = l.strip().split(' ')
  frames.append(elements)
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import cv2
i=0
p=0
j=1
rootdir='kinetics400test_images/kinetics400test_images'
rootdir2 = 'kinetics400test_audio/kinetics400test_audio'
# plt.figure(figsize=(40, 10))
for file in os.listdir(rootdir):
  d = os.path.join(rootdir, file)
  e = os.path.join(rootdir2, file)
  if os.path.isdir(d):
      fig = plt.figure(figsize=(50, 10))
      j=1
      m_idx = []
      for x in range(len(frames[i])):
        if os.path.exists(d+"/"+frames[i][x].zfill(5)+".jpg"):
          img = cv2.imread(d+"/"+frames[i][x].zfill(5)+".jpg")
        else:
          m_idx.append(x)
          plt.subplot(2,10,j)
          plt.axis("off")
          j+=1
          continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(2,10,j)
        if frames_to_select['selections'][p][x][0]:
          plt.imshow(img)
        else:
          plt.imshow(img,alpha=0.3)
        plt.axis("off")
        j+=1
        # plt.show()
      y, sr = librosa.load(e+'.wav')
      q = len(y)//10
      k=0
      for x in range(10):
        if x in m_idx:
          plt.subplot(2,10,j)
          plt.axis("off")
          j+=1
          continue
        else:
          plt.subplot(2,10,j)
          if frames_to_select['selections'][p][x][1]:
            librosa.display.waveplot(y[int(k):int(k+q)], sr=sr)
          else:
            librosa.display.waveplot(y[int(k):int(k+q)], sr=sr,alpha=0.3)
        plt.axis("off")
        k=k+q+1
        j+=1
  plt.suptitle(file, fontsize=48)
  fig.patch.set_linewidth(5)
  fig.patch.set_edgecolor('black')
  plt.show()
  plt.close()
  i+=1
  p+=1
# plt.show()
```
